## **2. Genera las columnas adicionales y luego expórtelas en un archivo synthetic_features.csv**

## **Temporada Alta**  

Para esta nueva variable, utilizaremos solo la columna de la variable "Fecha-I", esta columna la dejaremos como DataFrame, luego definimos los datos de este DataFrame como tipo "datetime" para utilizar el Framework de manipulación de datos temporales **datetime**, finalmente utilizamos la función *loc[]* para reemplazar por 1 y 0 los rangos de fechas solicitado, luego renombraremos la columna del DataFrame con el nombre **temporada_alta**.

```python
import datetime
import numpy as np
import pandas as pd


temporada_alta = pd.DataFrame(DUMMY_ORDINAL_ENCODED_DATA["Fecha-I"], columns=["Fecha-I"])

temporada_alta = temporada_alta.apply(pd.to_datetime)

temporada_alta.loc[((temporada_alta["Fecha-I"] >= datetime.datetime(2017, 12, 15)) & 
                    (temporada_alta["Fecha-I"] <= datetime.datetime(2018,  3,  3))) | 
                   ((temporada_alta["Fecha-I"] >= datetime.datetime(2017, 7, 15)) & 
                   (temporada_alta["Fecha-I"] <= datetime.datetime(2017, 7, 31))) | 
                   ((temporada_alta["Fecha-I"] >= datetime.datetime(2017, 9, 11)) & 
                    (temporada_alta["Fecha-I"] <= datetime.datetime(2017, 9, 30))) , 'Fecha-I']  = '1'

temporada_alta.loc[(temporada_alta["Fecha-I"] != "1") ,'Fecha-I'] = '0'

temporada_alta = temporada_alta.rename(columns={'Fecha-I':'temporada_alta'})
temporada_alta
```

## **Dif_min**  

Para esta variable utilizaremos las variables "Fecha-I" y "Fecha-O", sus datos los transformaremos a tipo datetime para luego restar la "Fecha-O" con "Fecha-I" y finalmente dividir por la función *Timedelta()* en donde su argumento será "1 min" para poder dejar los resultados en minutos.

```python
DUMMY_ORDINAL_ENCODED_DATA["Fecha-I"] = DUMMY_ORDINAL_ENCODED_DATA["Fecha-I"].apply(pd.to_datetime)

DUMMY_ORDINAL_ENCODED_DATA["Fecha-O"] = DUMMY_ORDINAL_ENCODED_DATA["Fecha-O"].apply(pd.to_datetime)

DUMMY_ORDINAL_ENCODED_DATA["Dif_min"] = (DUMMY_ORDINAL_ENCODED_DATA["Fecha-O"] - DUMMY_ORDINAL_ENCODED_DATA["Fecha-I"]) / pd.Timedelta('1 min')
Dif_min= DUMMY_ORDINAL_ENCODED_DATA["Dif_min"]
Dif_min
```

## **Atraso_15**  

Para esta variable utilizaremos la variable nueva "Dif_min" y la definiremos como DataFrame, luego através de la función *loc[]* actualizaremos los valores con 0 o 1 según el umbral dado en el desafío. Finalmente, renombraremos como "atraso_15" la columna del DataFrame ya que estaba como "Dif_min".

```python
atraso_15 = pd.DataFrame(DUMMY_ORDINAL_ENCODED_DATA["Dif_min"], columns=["Dif_min"])

atraso_15.loc[atraso_15["Dif_min"] > 15, "Dif_min"] = "1"
atraso_15.loc[(atraso_15["Dif_min"] != "1") ,'Dif_min'] = '0'
atraso_15 = atraso_15.rename(columns={'Dif_min':'atraso_15'})

atraso_15
```

**Exportación de nuevas variables a archivo .csv**

```python
synthetic_features = pd.concat([temporada_alta, atraso_15, Dif_min], axis = 1)
synthetic_features.to_csv("synthetic_features.csv",index=False)
```

## **3. ¿Cómo se compone la tasa de atraso por destino, aerolínea, mes del año, día de la semana, temporada, tipo de vuelo? ¿Qué variables esperarías que más influyeran en predecir atrasos?**  

Esta pregunta no la voy a responder porque usaré una técnica de selección de características para obtener las características mas importantes. Al final del notebook podrán ver cuáles fueron las más importantes.

**Ahora, concatenaremos las variables nuevas con la data que tenemos, excepto la variable "atraso_15", ya que muestra si un vuelo se retrasó o no. La variable "atraso_15" se dejará como TARGET por lo mencionado anteriormente.  
Así, estamos frente a un problema de Machine Learning de tipo Supervisado en donde la tarea es clasificar si un vuelo llega atrasado(1) o no(0).**

```python
Data_para_escalar = pd.concat([DUMMY_ORDINAL_ENCODED_DATA, atraso_15, temporada_alta], axis = 1)
Data_para_escalar
```

## Target  
El target será la característica "atraso_15" dado que muestra si el avion llegó o no atrasado. En esta parte separaremos el dataframe y definiremos **X** como las features e **y** como el target.  
El valor si el avión llegó atrasado es de 1.  
El valor se el avion NO llegó atrasado es 0.  

En las features no se considerarán las variables "Fecha-I" y "Fecha-O" para generar un modelo que prediga el atraso independientemente de las fechas de vuelo. Tampoco se considerará la variable "SIGLAORI" por lo mencionado en la pregunta 1.

```python
#Dividimos dataset en features y label

#FEATURES
X = Data_para_escalar[Data_para_escalar.columns.difference(['atraso_15',"Fecha-I","Fecha-O","SIGLAORI"])]

#LABELS
y = pd.DataFrame(Data_para_escalar.atraso_15, columns=["atraso_15"])
```

## Escalar Data   

Ahora escalaremos los datos para que el modelo tenga un mejor rendimiento. Utilizaremos el método **MinMaxScaler()** para realizar esta tarea, dejando todos los valores de la data en el rango [0,1]  

FUENTE: https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html

```python
from sklearn.preprocessing import MinMaxScaler

X_escalado = MinMaxScaler().fit_transform(X)
X_escalado = pd.DataFrame(X_escalado, columns = X.columns)
X_escalado
```

**Ahora, una persona de la industria podría decir según su experiencia e intuición cuál variable puede tener mayor importancia, pero existen distintas técnicas para seleccionar características. Para este caso utilizaremos la técnica *Mutual Information*, técnica supervisada que mide la dependencia entre dos variables aleatorias, así, mide la dependencia de cada feature con el target. El valor es 0 si no hay dependencia y mientras mayor el valor, mayor la dependencia.**  

FUENTE: https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.mutual_info_classif.html

En la implementación graficaremos las características y su relevancia, como son muchas el gráfico no dice mucho, pero luego mostraremos y nos quedaremos con las 15 variables de mayor puntaje.

```python
from sklearn.feature_selection import mutual_info_classif
import matplotlib.pyplot as plt
%matplotlib inline

importances = mutual_info_classif(X_escalado,y)
feat_importances = pd.Series(importances, X_escalado.columns[0:len(X_escalado.columns)])

plt.figure(figsize=(6,14))
feat_importances.plot(kind = "barh",color = "teal")

plt.show()
```

```python
umbral = 15  # cantidad de features mas importantes según su puntaje, puede variar.
caract_mayor_puntaje = []

for score, f_name in sorted(zip(importances, X_escalado.columns), reverse=True)[:umbral]:
    print(f_name, score)
    caract_mayor_puntaje.append(f_name)

#Acá mostramos las 15 features mas importantes y su respectivo puntaje    
caract_y_puntaje = X_escalado[caract_mayor_puntaje]
```

**Ahora actualizamos la data que teníamos con solo las features importantes según el método señalado.**

```python
data_final=pd.DataFrame(X_escalado[(caract_y_puntaje.columns)])
data_final
```

## Desbalance de clases  

Siempre es bueno revisar si las clases del target estan balanceadas, ya que de no ser así, al momento de entrenar un modelo de ML este puede resultar sesgado.  
La literatura sugiere que si la proporción de la clase minoritaria está entre el 1% y el 20%, el desbalance es de grado Medio.  
Ahora graficaremos el target para ver el grado de desbalance y, con la funcion *value_counts()* obtener cuántos registros hay de cada clase para calcular el porcentaje de la clase minoritaria.

```python
import seaborn as sns

sns.countplot(y["atraso_15"])
```

```python
y["atraso_15"].value_counts() #total 68206
```

**El porcentaje de cada clase es 18% clase 1 y 82% clase 0, dado el valor del porcentaje de la clase minoritaria, estamos frente a un grado medio de desbalance de clases, por tanto tenemos que resolver este problema. Una técnica que nos ayuda en esta tarea es SMOTE del Framework Imbalanced Learn, el cuál a través de un algoritmo de k-vecinos más cercanos, genera datos sintéticos hasta balancear las clases.**  

FUENTE: https://imbalanced-learn.org/stable/over_sampling.html#mathematical-formulation

```python
from imblearn.over_sampling import SMOTE

X_resample,y_resample=SMOTE().fit_resample(data_final,y.values.ravel())

y_resample=pd.DataFrame(y_resample,columns=y.columns)
X_resample=pd.DataFrame(X_resample,columns=data_final.columns)
```

```python
sns.countplot(y_resample["atraso_15"])
```

```python
#Revisamos si los datos estan balanceados, el algoritmo igualó la clase mayoritaria.

y_resample.atraso_15.value_counts()
```

**Con lo anterior terminamos el proceso de Feature Engineering, ahora dividimos la data y el target en entrenamiento y prueba para su posterior uso en un modelo de clasificación.**

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X_resample,y_resample,test_size=0.3,random_state=101)
```

```python
#Cambiamos el tipo de dato a entero en y_test ya que era de tipo object.
y_test = y_test.astype(int)
```

## **4. Entrena uno o varios modelos (usando el/los algoritmo(s) que prefieras) para estimar la probabilidad de atraso de un vuelo.Siéntete libre de generar variables adicionales y/o complementar con variables externas.**  

Las variables adicionales o externas se generaron en la etapa de Feature Engineering.  
El modelo que se eligió es SVM, el cual a través de hiperplanos realiza la clasificación.

```python
from sklearn.svm import SVC

svc=SVC()
svc.fit(x_train,y_train.values.ravel())
y_pred_SVC=svc.predict(x_test)
```

```python
#Cambiamos el tipo de dato a entero en y_pred_SVC ya que era de tipo object.

y_pred_SVC = y_pred_SVC.astype(int)
```
